# Read MDF files 

Following [Reading MS Access MDB Files on Mac](https://medium.com/@wenyu.z/reading-ms-access-mdb-files-on-mac-969a176baa7a).

Pre-requsite: `mdbtools` and/or `meza`.

```python
def show_tables(path='<file_name>.mdb'):
    tables = subprocess.check_output(["mdb-tables", path])
    return tables.decode().split()

def show_data(path='<file_name>.mdb', table='<table_name>'):
    tables = subprocess.check_output(["mdb-export", path, table])
    return tables.decode().split('\n')
 
def convert_df(path, table):
    d = show_data(path, table)
    columns = d[0].split(',')
#     data = [i.split(',')[:len(columns)] for i in d[1:]]
    # use more smarty way to separate the column data
    data = [re.findall(r'(?:,|\n|^)("(?:(?:"")*[^"]*)*"|[^",\n]*|(?:\n|$))', i)[:len(columns)] for i in d[1:]]
    df = pd.DataFrame(columns=columns, data=data)
    return df
```

```python
tables = show_tables(fname)
tables
```

```python
df = convert_df(fname, table=tables[0])
```
